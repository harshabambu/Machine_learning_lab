```python
import numpy as np
features=np.array([[0,0],[0,1],[1,0],[1,1]])
labels=np.array([0,0,0,1])
```

```python
epochs = int(input("Enter the number of epochs: "))
threshold = float(input("Enter the threshold: "))
learning_rate = float(input("Enter the learning rate: "))

# Initialize weights
a = []
a.append(float(input("Enter the weight for x0: ")))
a.append(float(input("Enter the weight for x1: ")))
w = np.array(a)

# Assuming features and labels are defined before running this code
for j in range(epochs):
    print("Epoch:", j)
    global_delta = 0

    for i in range(features.shape[0]):
        actual = labels[i]
        instance = features[i]
        x0 = instance[0]
        x1 = instance[1]
        sum_unit = x0 * w[0] + x1 * w[1]
        
        if sum_unit > threshold:
            fire = 1
        else:
            fire = 0

        delta = actual - fire
        global_delta += abs(delta)
        
        print(f"Prediction: {fire}, whereas actual was {actual} (Error: {delta})")
        
        w[0] = w[0] + delta * learning_rate
        w[1] = w[1] + delta * learning_rate

    print("--------------------")

    if global_delta == 0:
        break

print("Final weights:", w)
```
