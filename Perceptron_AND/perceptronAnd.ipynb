```python
import numpy as np

# Step-1: Initialize Input and Output Variables
features = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
labels = np.array([0, 0, 0, 1])  # AND Gate Outputs

# Step-2: Initialize the Network Parameters
epochs = int(input('Enter the Epochs: '))
threshold = float(input('Enter the Threshold: '))
learning_rate = float(input('Enter the Learning Rate: '))

# Step-3: Initialize Weights
a = []
a.append(float(input('Enter the weight for x0: ')))
a.append(float(input('Enter the weight for x1: ')))
w = np.array(a)

# Step-4: Train the Perceptron
for j in range(epochs):  # Loop through epochs
    print("Epoch", j)
    global_delta = 0  # Keeps track of errors in this epoch

    for i in range(features.shape[0]):  # Loop through each training example
        actual = labels[i]  # Expected output
        instance = features[i]  # Input values
        x0, x1 = instance[0], instance[1]  # Extract input values

        print("X0:", x0, "X1:", x1)

        # Compute the weighted sum
        sum_unit = x0 * w[0] + x1 * w[1]
        print("Sum Value:", sum_unit)

        # Activation function
        if sum_unit > threshold:
            fire = 1
        else:
            fire = 0

        # Calculate error
        delta = actual - fire
        global_delta += abs(delta)

        print("Prediction:", fire, "whereas actual was", actual, "(error:", delta, ")")

        # Update weights
        w[0] = w[0] + delta * learning_rate
        w[1] = w[1] + delta * learning_rate

        print("--------------------")

    # If no error, stop training
    if global_delta == 0:
        break

print("Final Weights:", w)
```
